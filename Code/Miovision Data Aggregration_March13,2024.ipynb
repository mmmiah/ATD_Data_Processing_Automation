```python
# import libraries
import pandas as pd
import numpy as np
import os
```

```python
# Directory containing Excel files
directory = 'C:/UC Berkeley Project Backup/PSIP_Phase_4_Analysis/Miovision_data/Data Processing Script/Miovision_data'
```

```python
aggregated_data = []
# Iterate over each Excel file
for filename in os.listdir(directory):
    if filename.endswith('.xlsx'):
        filepath = os.path.join(directory, filename)
        try:
            with pd.ExcelFile(filepath) as xls:
                # Read Summary sheet
                summary_df = pd.read_excel(xls, 'Summary')

                # Check if condition is met in summary sheet
                if (summary_df.iloc[2, 1] == 'All Processed Legs & Movements'):
                    location = summary_df.iloc[7, 1]  # Location is in cell B9
                    latlon = summary_df.iloc[8, 1]
                    latlon = latlon.split(',')
                    lat = latlon[0]
                    lon = latlon[1]
                    # Read Pedestrians sheet
                    pedestrians_df = pd.read_excel(xls, 'Pedestrians', header=2)

                    # Sum up all columns' values by first column starting from row 4
                        # Sum up all columns' values by first column starting from row 4
                    pedestrians_df['total_Counts'] = pedestrians_df.iloc[:, 1:].sum(axis=1)
                    pedestrians_df['location'] = location
                    pedestrians_df['lat'] = lat
                    pedestrians_df['long'] = lon

                    # Append to aggregated data list
                    aggregated_data.append(pedestrians_df)
        except Exception as e:
            print(f"Error processing {filename}: {str(e)}")

# Concatenate data into a single DataFrame
result_df = pd.concat(aggregated_data)
result_df
```
