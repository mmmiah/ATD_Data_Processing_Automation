```python
import pandas as pd
import geopandas as gpd
```

# lets combine ped data

```python
# Read the D4 ped data
d4_ped=pd.read_csv('Z:/AT Data/Tableau Database/Archive/For Mintu/D3/FOR COMPILATION/Ped/Tableau_D4_Pedestrian.csv')
# drop the duplicates
d41_ped=d4_ped.drop_duplicates(subset=['Loc_id', 'District', 'Intersection', 'Year', 'Month', 'Day',
       'Weekday_Type', 'Date', 'Hour', 'Mode', 'Direction', 'Count'])
d41_ped.head()
```

```python
d4_ped.shape,d41_ped.shape
```

```python
d4_ped.columns
```

```python
d3_ped=pd.read_csv('Z:/AT Data/Tableau Database/Archive/For Mintu/D3/FOR COMPILATION/Ped/Tableau_D3_Pedestrian.csv')
d3_ped.head()
```

```python
d3_ped.shape,d41_ped.shape
```

```python
d3_ped.dtypes,d41_ped.dtypes
```

```python
# lets combine these two datasets
ped_D3_D4_data=pd.concat([d3_ped,d41_ped],axis=0)
ped_D3_D4_data.shape
```

```python
ped_D3_D4_data.isnull().sum()
```

```python
ped_D3_D4_data.to_csv('Z:/AT Data/Tableau Database/Archive/For Mintu/D3/FOR COMPILATION/Ped/combined_D3_D4_ped.csv',index=False)
```

# lets combine bike data

```python
# Read the D4 bikr data
d4_bike=pd.read_csv('Z:/AT Data/Tableau Database/Archive/For Mintu/D3/FOR COMPILATION/Bike/Tableau_D4_Bicycle.csv')
# drop the duplicates
d41_bike=d4_bike.drop_duplicates(subset=['Loc_id', 'District', 'Intersection', 'Year', 'Month', 'Day',
       'Weekday_Type', 'Date', 'Hour', 'Mode', 'Direction', 'Count'])
d41_bike['Count']=d41_bike['Count'].astype('int64')
d41_bike.head()
```

```python
d41_bike.isnull().sum()
```

```python
d4_bike.shape,d41_bike.shape
```

```python
d3_bike=pd.read_csv('Z:/AT Data/Tableau Database/Archive/For Mintu/D3/FOR COMPILATION/Bike/Tableau_D3_Bicycle.csv')
d3_bike.head()
```

```python
d3_bike.isnull()
```

```python
d3_bike.shape,d41_bike.shape
```

```python
d3_bike.dtypes,d41_bike.dtypes
```

```python
# lets combine these two datasets
bike_D3_D4_data=pd.concat([d3_bike,d41_bike],axis=0)
bike_D3_D4_data = bike_D3_D4_data.dropna()
bike_D3_D4_data.shape
```

```python
bike_D3_D4_data.isnull().sum()
```

```python
bike_D3_D4_data.to_csv('Z:/AT Data/Tableau Database/Archive/For Mintu/D3/FOR COMPILATION/Bike/combined_D3_D4_bike.csv',index=False)
```
