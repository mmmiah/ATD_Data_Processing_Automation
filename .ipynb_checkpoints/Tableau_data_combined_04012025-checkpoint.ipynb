```python
import pandas as pd
```

```python
# read the bike file
f1=pd.read_csv('Z:/AT Data/Data_to_be_processed/Bicycle Location ID/D6_bike_data_for_Tableau.csv')
f1['Count_Type']='Continuous'
f1["Date"] = pd.to_datetime(f1["Date"]).dt.strftime("%m/%d/%Y %H:%M")
f2=pd.read_csv('Z:/AT Data/Data_to_be_processed/Bicycle Location ID/Tableau_Bicycle.csv')
# concate
com_file=pd.concat([f1,f2],axis=0)
com_file=com_file.drop(['Unnamed: 0'],axis=1)
com_file.to_csv('Z:/AT Data/Processed_data/Combined_data/Tableau_bike_data_04022025.csv',index=False)
```

```python
# read the ped file
fp1=pd.read_csv('Z:/AT Data/Data_to_be_processed/Pedestrian Location ID/D6_ped_data_for_Tableau.csv')
fp1['Count_Type']='Continuous'
fp1["Date"] = pd.to_datetime(fp1["Date"]).dt.strftime("%m/%d/%Y %H:%M")
fp2=pd.read_csv('Z:/AT Data/Data_to_be_processed/Pedestrian Location ID/Tableau_Pedestrian.csv')
# concate
com_filep=pd.concat([fp1,fp2],axis=0)
com_filep=com_filep.drop(['Unnamed: 0'],axis=1)
com_filep.to_csv('Z:/AT Data/Processed_data/Combined_data/Tableau_ped_data_04022025.csv',index=False)
```
